```python

import pandas as pd

data = pd.read_csv("train.csv")
data.drop(data.columns[0],axis = 1,inplace=True)
data.head()
```

```python
import seaborn as sb

#checked for outliers
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#checked distribution of data
data.describe()
```

```python
#missing value check
data.isnull().sum()
```

```python
# If data is balanced or not
data['labels'].value_counts()
```

```python
X = data.iloc[:,0:-1]
y = data.iloc[:,-1]

val_data = pd.read_csv("validation.csv")
val_data.drop(val_data.columns[0],axis = 1,inplace = True)
val_data.head()

#y.head()
```

```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.25, random_state=0)
x_validation = val_data.iloc[:,0:-1]
y_validation = val_data.iloc[:,-1]
NB = MultinomialNB()
NB.fit(X_train, y_train)
y_pred = NB.predict(X_test)
print("Naive Bayes cross validation Accuracy:",metrics.accuracy_score(y_test, y_pred))
y_pred = NB.predict(x_validation)
print("Naive Bayes Accuracy:",metrics.accuracy_score(y_validation, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing 

X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.25, random_state=0)
X_train = preprocessing.normalize(X_train)
X_test = preprocessing.normalize(X_test)
x_validation = val_data.iloc[:,0:-1]
x_validation = preprocessing.normalize(x_validation)
y_validation = val_data.iloc[:,-1]
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
y_pred = knn.predict(x_validation)
print("Accuracy:",metrics.accuracy_score(y_validation, y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import preprocessing 

X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.25, random_state=0)
X_train = preprocessing.normalize(X_train)
X_test = preprocessing.normalize(X_test)
x_validation = val_data.iloc[:,0:-1]
x_validation = preprocessing.normalize(x_validation)
y_validation = val_data.iloc[:,-1]
clf = AdaBoostClassifier(n_estimators=80,learning_rate=1,random_state=0)
model = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
y_pred = clf.predict(x_validation)
print("Accuracy:",metrics.accuracy_score(y_validation, y_pred))
```

```python
import numpy as np

corr_matrix = data.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
print(to_drop)
for val in to_drop:
    data.drop(data.columns[int(val)], axis=1)
for val in to_drop:
    val_data.drop(data.columns[int(val)], axis=1)
print("done")

x_train_drop = data.iloc[:,0:-1]
y = data.iloc[:,-1]

x_validation_drop = val_data.iloc[:,0:-1]
y = data.iloc[:,-1]
```
